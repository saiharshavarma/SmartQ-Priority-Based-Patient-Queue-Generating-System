```python
import random
import csv

def generate_dataset(num_samples):
    dataset = []
    for i in range(num_samples):
        age = random.randint(18, 80)
        gender = random.choice(["male", "female"])
        past_history = random.choice(["none", "diabetes", "hypertension", "both"])
        current_symptoms = random.choice(["none", "fever", "cough", "sore throat", "difficulty breathing", "shortness of breath",
                                          "fever,cough,sore throat", "fever,difficulty breathing", "cough,difficulty breathing",
                                          "fever,cough,difficulty breathing"])
        score = 0
        
        if age >= 60:
            score += 2
        elif age >= 40:
            score += 1
        
        if gender == "female":
            score += 1
        
        if "diabetes" in past_history:
            score += 2
        elif "hypertension" in past_history:
            score += 2
        
        if "difficulty breathing" in current_symptoms or "shortness of breath" in current_symptoms:
            score += 4
        elif "fever" in current_symptoms or "cough" in current_symptoms or "sore throat" in current_symptoms:
            score += 1
        
        if score <= 2:
            criticality = 1
        elif score <= 4:
            criticality = 2
        elif score <= 6:
            criticality = 3
        elif score <= 8:
            criticality = 4
        else:
            criticality = 5
        
        dataset.append([age, gender, past_history, current_symptoms, criticality])
        
    return dataset

# Generate a dataset with 1000 samples
dataset = generate_dataset(1000)

# Write the dataset to a CSV file
with open('dataset.csv', 'w', newline='') as file:
    writer = csv.writer(file)
    writer.writerow(["age", "gender", "past_history", "current_symptoms", "criticality"])
    for data in dataset:
        writer.writerow(data)
```
