```python
dataset = pd.read_csv('/content/drive/MyDrive/MAKEATHON2K23/Dataset/criticality_dataset.csv')
```

```python
dataset.head()
```

```python
gender_list = []
diabetes_list = []
hypertension_list = []
fever_list = []
cough_list = []
sore_throat_list = []
difficulty_breathing_list = []
shortness_of_breath_list = []
```

```python
for i in dataset['gender']:
  if i == 'female':
    gender_list.append(0)
  else:
    gender_list.append(1)
```

```python
for i in dataset['diabetes_history']:
  if i == 'no':
    diabetes_list.append(0)
  else:
    diabetes_list.append(1)
```

```python
for i in dataset['hypertension_history']:
  if i == 'no':
    hypertension_list.append(0)
  else:
    hypertension_list.append(1)
```

```python
for i in dataset['fever']:
  if i == 'no':
    fever_list.append(0)
  else:
    fever_list.append(1)
```

```python
for i in dataset['cough']:
  if i == 'no':
    cough_list.append(0)
  else:
    cough_list.append(1)
```

```python
for i in dataset['sore_throat']:
  if i == 'no':
    sore_throat_list.append(0)
  else:
    sore_throat_list.append(1)
```

```python
for i in dataset['difficulty_breathing']:
  if i == 'no':
    difficulty_breathing_list.append(0)
  else:
    difficulty_breathing_list.append(1)
```

```python
for i in dataset['shortness_of_breath']:
  if i == 'no':
    shortness_of_breath_list.append(0)
  else:
    shortness_of_breath_list.append(1)
```

```python
dataset = dataset.drop(['gender',	'diabetes_history',	'hypertension_history',
                        'fever',	'cough',	'sore_throat',	'difficulty_breathing',
                        'shortness_of_breath'], axis=1)
```

```python
dataset['gender'] = gender_list
dataset['diabetes_history'] = diabetes_list
dataset['hypertension_history'] = hypertension_list
dataset['fever'] = fever_list
dataset['cough'] = cough_list
dataset['sore_throat'] = sore_throat_list
dataset['difficulty_breathing'] = difficulty_breathing_list
dataset['shortness_of_breath'] = shortness_of_breath_list
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2)
```

```python
from sklearn.tree import DecisionTreeClassifier
clf = DecisionTreeClassifier()
clf = clf.fit(X_train,y_train)
y_pred = clf.predict(X_test)
```

```python
y_train_pred = clf.predict(X_train)
```

```python
y_pred = clf.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score
print('Train Accuracy: {}%'.format(round(accuracy_score(y_train, y_train_pred), 4)*100))
```

```python
from sklearn.metrics import accuracy_score
print('Test Accuracy: {}%'.format(round(accuracy_score(y_test, y_pred), 4)*100))
```
